```python
def exponential_func(x, a, b):
    return a * np.exp(b * x)


x = np.array(x)
y = np.array(y)

y_log = np.log(y)

# Perform linear regression to estimate parameters a and b
A = np.vstack([x, np.ones_like(x)]).T
a_log, b_log = np.linalg.lstsq(A, y_log, rcond=None)[0]

# Calculate parameters a and b for the exponential model
a = np.exp(b_log)
b = a_log

print(f"a = {a}, b = {b}")

x_fit = np.linspace(1, len(y), 100)
y_fit = exponential_func(x_fit, a, b)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Bag Production', linestyle='-')
plt.plot(x_fit, y_fit, 'r-', label='Exponential Fit')
plt.xlabel('Time Period')
plt.ylabel('Bag Production')
plt.title('Exponential Fit to Bag Production Data')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
